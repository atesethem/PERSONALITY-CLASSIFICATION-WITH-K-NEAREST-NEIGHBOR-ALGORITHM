```python
def prediction(c,k):
    for i in c[:,:k]:
        global mypersonality        
        mypersonality.append(st.mode(i))
    mypersonality = np.array(mypersonality)
```

```python
def creatingconfusionmatrix(personality1):
    confusion_matrix = np.zeros([16, 16])
    np.add.at(confusion_matrix, (personality1 - 1, mypersonality - 1), 1)
    recall = np.sum(np.diag(confusion_matrix) / np.sum(confusion_matrix,1))/16
    precision = np.sum(np.diag(confusion_matrix) / np.sum(confusion_matrix,0))/16
    accuracy = np.sum(np.diag(confusion_matrix)) / np.sum(confusion_matrix)
    print("Recall: ", recall)
    print("Precision: ", precision)
    print("Accuracy: ", accuracy)
```
